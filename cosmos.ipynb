```python
import os
from PyPDF2 import PdfReader

def extract_text_from_pdf(file_path):
    """
    Extract text from a given PDF file.
    :param file_path: Path to the PDF file.
    :return: Extracted text from the PDF.
    """
    reader = PdfReader(file_path)
    text = ""
    for page in reader.pages:
        text += page.extract_text()
    return text

def extract_book_name(file_path):
    """
    Extract the book title from the PDF file name.
    Assumes the format "(요약본)책 제목".
    :param file_path: Path to the PDF file.
    :return: Extracted book title.
    """
    # Get the file name without the path
    file_name = os.path.basename(file_path)

    # Remove the prefix in parentheses (e.g., "(요약본)")
    if ")" in file_name:
        book_title = file_name.split(")", 1)[1].strip()
    else:
        book_title = file_name  # If no parentheses, use the full file name

    return book_title

# 다운로드된 파일 예시
current_directory = os.getcwd()
download_folder = os.path.join(current_directory, "download")

# 첫 번째 PDF 파일 경로 가져오기
downloaded_files = os.listdir(download_folder)
if downloaded_files:
    first_pdf_path = os.path.join(download_folder, downloaded_files[0])

    # 책 제목 추출
    pdf_name = extract_book_name(first_pdf_path)
    print(f"Extracted Book Title: {pdf_name}")

    # PDF 텍스트 추출
    pdf_text = extract_text_from_pdf(first_pdf_path)
    print("Extracted Text from PDF:")
    print(pdf_text)
else:
    print("No PDF files found in the download folder.")
```

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
import csv

# 환경 변수 로드
load_dotenv()
client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY")
)

# ChatGPT 요청 함수
def extract_genres(plot_summary):
    try:
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",  # 또는 "gpt-4"를 사용할 수 있습니다.
            messages=[
                {"role": "system", "content": "You are a helpful assistant that extracts genres from book summaries."},
                {"role": "user", "content": f"Extract up to 3 genres from the following plot summary:\n{plot_summary}"}
            ]
        )
        genres = response.choices[0].message.content
        return [genre.strip() for genre in genres.split(",")][:3]
    except Exception as e:
        print(f"ChatGPT API 호출 중 오류가 발생했습니다: {e}")
        return None

def append_to_csv(file_name, column_names, data):
    """
    데이터를 CSV 파일에 추가합니다.
    :param file_name: CSV 파일 이름
    :param column_names: CSV 컬럼 이름 목록
    :param data: 추가할 데이터 (딕셔너리 형태)
    """
    file_exists = os.path.isfile(file_name)
    with open(file_name, mode='a', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=column_names)
        if not file_exists:
            writer.writeheader()
        writer.writerow(data)

# CSV 파일 및 컬럼 설정
csv_file_name = "books.csv"
columns = ["Book title", "Author", "Book genres (Freebase ID:name tuples)", "Plot summary"]

# 책 정보 (예제)
book_title = book_name
author = author_name
plot_summary = cleaned_summary

# OpenAI API를 사용하여 장르 추출
genres = extract_genres(plot_summary)
genres_str = ", ".join(genres)  # 장르 리스트를 문자열로 변환

# 데이터를 CSV에 추가
book_data = {
    "Book title": book_title,
    "Author": author,
    "Book genres (Freebase ID:name tuples)": genres_str,
    "Plot summary": plot_summary
}
append_to_csv(csv_file_name, columns, book_data)

print(f"다음 데이터가 '{csv_file_name}'에 저장되었습니다:")
print(book_data)
```
