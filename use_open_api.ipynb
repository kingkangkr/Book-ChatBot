```python
import requests
from bs4 import BeautifulSoup

# 책 제목 리스트
unique_book_titles = ['불편한 편의점', '아버지의 해방일지', '달러구트 꿈 백화점', '어서오세요, 휴남동 서점입니다']

# 위키피디아 URL 생성 및 HTML 가져오기
base_url = "https://ko.wikipedia.org/wiki/"
results = {}

for title in unique_book_titles:
    # 책 제목을 URL에 맞게 인코딩
    url = base_url + requests.utils.quote(title)
    print(f"Fetching: {url}")
    try:
        response = requests.get(url)
        response.raise_for_status()  # HTTP 에러 발생 시 예외 처리
        html_content = response.text
        
        # BeautifulSoup으로 HTML 파싱
        soup = BeautifulSoup(html_content, 'html.parser')
        
        # HTML에서 텍스트 추출
        raw_text = soup.get_text(separator="\n", strip=True)
        
        # 불필요한 키워드 제거 (예: 메뉴, 툴바 등)
        useful_text = []
        for line in raw_text.split("\n"):
            if not any(keyword in line for keyword in ["주 메뉴", "PDF로 다운로드", "본문으로 이동", "사이드바로 이동", "둘러보기", "숨기기", "사용자 모임", "도움말", "정책과 지침", "대문", "최근 바뀜",
                                                       "요즘 화제", "로그인", "관리 요청", "편집 안내", "검색", "기부", "임의의 문서로",
                                                       "소개방", "보이기", "계정 만들기", "개인 도구",
                                                       "사랑방", "소개", "질문방", "로그아웃한 편집자를 위한 문서", "기여",
                                                       "더 알아보기", "각주", "언어 추가"]):
                useful_text.append(line)
        
        # 필터링된 텍스트를 저장
        clean_text = "\n".join(useful_text)
        results[title] = clean_text
        print(f"Cleaned text for {title}:\n{clean_text[:500]}...\n")
    except requests.exceptions.HTTPError as e:
        print(f"HTTP Error for {title}: {e}")
    except Exception as e:
        print(f"Error for {title}: {e}")

# 결과 출력
for title, clean_text in results.items():
    print(f"Cleaned text for {title}:\n{clean_text[:500]}...\n")
    break
```

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

# 환경 변수 로드
load_dotenv()
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
openai_client = OpenAI(api_key=OPENAI_API_KEY)


# 모델 설정
model = "gpt-4o-mini"

# ChatGPT 요청 함수
def chatgpt_generate(query):
    messages = [
        {
            "role": "system",
            "content": "You are a helpful assistant. Analyze the following text and determine if it is related to a novel."
        },
        {
            "role": "user",
            "content": query
        }
    ]
    response = openai_client.chat.completions.create(model=model, messages=messages)
    answer = response.choices[0].message.content
    return answer

# 전처리된 텍스트 전달 및 판단
for title, clean_text in results.items():  # results는 전처리된 텍스트 딕셔너리
    query = f"""
    The following text was extracted from a webpage. Please analyze the text and determine if it describes a novel. 
    Respond with "Novel" if it is related to a novel, otherwise respond with "Not a Novel".

    Title: {title}

    Text:
    {clean_text}
    """
    try:
        answer = chatgpt_generate(query)
        print(f"Response for '{title}': {answer}")
    except Exception as e:
        print(f"Error processing '{title}': {e}")
```

```python
# '도서명정보' 열에서 책 제목만 추출하고 중복 제거
book_titles = data['도서명정보'].str.split(':').str[0]  # ':'로 나눈 후 첫 번째 부분만 추출
unique_book_titles = book_titles.drop_duplicates().tolist()  # 중복 제거 후 리스트로 변환

print(unique_book_titles)
print(len(unique_book_titles))


# 위키피디아 URL
base_url = "https://ko.wikipedia.org/wiki/"
results = {}
success_titles = []  #  검색 성공한 책 제목
failed_titles = []   #  검색 실패한 책 제목

for title in unique_book_titles:
    # 책 제목을 URL에 맞게 인코딩
    url = base_url + requests.utils.quote(title)
    print(f"Fetching: {url}")
    
    try:
        response = requests.get(url)
        response.raise_for_status()  # HTTP 에러 발생 시 예외 처리
        html_content = response.text
        
        # BeautifulSoup으로 HTML 파싱
        soup = BeautifulSoup(html_content, 'html.parser')
        
        #  검색 성공한 책 저장
        results[title] = soup
        success_titles.append(title)
        print(f"Fetched HTML for {title}")
    
    except requests.exceptions.HTTPError as e:
        print(f"HTTP Error for {title}: {e}")
        failed_titles.append(title)  #  검색 실패한 책 저장
    except Exception as e:
        print(f"Error for {title}: {e}")
        failed_titles.append(title)  #  검색 실패한 책 저장

#  검색 결과 요약
total_books = len(unique_book_titles)
success_count = len(success_titles)
failed_count = len(failed_titles)

success_ratio = (success_count / total_books) * 100
failed_ratio = (failed_count / total_books) * 100

print("\n 검색 결과 요약 ")
print(f"전체 도서 수: {total_books}")
print(f"검색 성공 도서 수: {success_count} ({success_ratio:.2f}%)")
print(f"검색 실패 도서 수: {failed_count} ({failed_ratio:.2f}%)")


```

```python


# 나무위키 URL
base_url = "https://namu.wiki/w/"
results = {}
success_titles2 = []  # 검색 성공한 책 제목
failed_titles2 = []   # 검색 실패한 책 제목

# User-Agent 설정 (나무위키는 User-Agent가 없으면 차단될 수 있음)
headers = {
    'User-Agent': 'Mozilla/5.0'
}

for title in unique_book_titles:
    # 책 제목을 URL에 맞게 인코딩
    url = base_url + requests.utils.quote(title)
    print(f"Fetching: {url}")
    
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # HTTP 에러 발생 시 예외 처리
        html_content = response.text
        
        # BeautifulSoup으로 HTML 파싱
        soup = BeautifulSoup(html_content, 'html.parser')
        
        # 페이지가 존재하는지 확인 (나무위키는 없는 페이지에 안내 메시지가 있음)
        if "문서가 존재하지 않습니다" in soup.text:
            print(f"문서 없음: {title}")
            failed_titles2.append(title)
        else:
            results[title] = soup
            success_titles2.append(title)
            print(f"문서 있음: {title}")
    
    except requests.exceptions.HTTPError as e:
        print(f"HTTP Error for {title}: {e}")
        failed_titles2.append(title)
    except Exception as e:
        print(f"Error for {title}: {e}")
        failed_titles2.append(title)

# 검색 결과 요약
total_books = len(unique_book_titles)
success_count = len(success_titles2)
failed_count = len(failed_titles2)

success_ratio = (success_count / total_books) * 100
failed_ratio = (failed_count / total_books) * 100

print("\n나무위키 검색 결과 요약")
print(f"전체 도서 수: {total_books}")
print(f"검색 성공 도서 수: {success_count} ({success_ratio:.2f}%)")
print(f"검색 실패 도서 수: {failed_count} ({failed_ratio:.2f}%)")

```

```python
# 위키피디아와 나무위키의 검색 성공 도서 비교
wikipedia_success_set = set(success_titles)   # 위키피디아 성공 도서 목록 (이미 저장되어 있다고 가정)
namuwiki_success_set = set(success_titles2)   # 나무위키 성공 도서 목록

# 교집합 (둘 다 검색 성공한 도서)
common_success_titles = wikipedia_success_set & namuwiki_success_set

# 비율 계산
common_count = len(common_success_titles)
total_wikipedia_success = len(wikipedia_success_set)

# 위키피디아 검색 성공 도서 대비 나무위키와 겹치는 비율
if total_wikipedia_success > 0:
    overlap_ratio = (common_count / total_wikipedia_success) * 100
else:
    overlap_ratio = 0

# 결과 출력
print("\n 위키피디아와 나무위키 검색 결과 비교 ")
print(f"위키피디아 검색 성공 도서 수: {total_wikipedia_success}")
print(f"나무위키 검색 성공 도서 수: {len(namuwiki_success_set)}")
print(f"두 플랫폼 모두에서 검색 성공한 도서 수: {common_count}")
print(f"위키피디아 성공 도서 중 나무위키에도 존재하는 비율: {overlap_ratio:.2f}%")

# 겹치는 도서 목록 출력
if common_success_titles:
    print("\n 두 플랫폼 모두에서 검색 성공한 도서 목록:")
    for title in common_success_titles:
        print(f"- {title}")
else:
    print("\n 두 플랫폼 모두에서 검색 성공한 도서가 없습니다.")
```
